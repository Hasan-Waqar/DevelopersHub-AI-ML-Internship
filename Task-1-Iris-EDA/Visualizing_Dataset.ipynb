<h1>Step 1: Import Necessary Libraries<h1>

First, we import the libraries required for our analysis.  
pandas: For data manipulation and loading.  
seaborn & matplotlib: For data visualization.  

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# This is a common setting to make plots look a bit nicer
sns.set_theme(style="whitegrid")
```

<h1>Step 2: Load the Dataset<h1>  

The Iris dataset is a classic in machine learning and is conveniently available in the seaborn library. We'll load it directly into a pandas DataFrame.

```python
# Load the Iris dataset from seaborn's built-in datasets
iris_df = sns.load_dataset('iris')
```

<h1>Step 3: Initial Data Inspection<h1>

Let's perform a basic check of the dataset to understand its structure. We will check:  
The dimensions (number of rows and columns).  
The names of the columns.  
The first few rows of data to get a feel for the values.  

```python
# 1. Print the shape of the dataset (rows, columns)
print("Shape of the dataset:", iris_df.shape)

# 2. Print the column names
print("\nColumn names:", iris_df.columns.tolist())

# 3. Print the first few rows using .head()
print("\nFirst 5 rows of the dataset:")
iris_df.head()
```

<h1>Step 4: Summary Statistics<h1>  

Now, we'll dive deeper by generating summary statistics.  
.info() provides a concise summary, including data types and checking for missing values.  
.describe() gives statistical details like mean, standard deviation, and quartiles for the numerical columns.  

```python
# Use .info() to get a concise summary of the dataframe
print("\nDataset Info:")
iris_df.info()

# Use .describe() to get summary statistics for numerical columns
print("\nSummary Statistics:")
iris_df.describe()
```

<h1>Step 5: Data Visualization<h1>

Visualization is key to understanding data. We will create three types of plots as requested:  
*   Scatter Plots: To see relationships between pairs of features. A pairplot is perfect for this.  

*   Histograms: To understand the distribution of each individual feature.  

*   Box Plots: To visualize the spread of data and identify potential outliers for each feature, broken down by species.  




<h2>5a. Pair-wise Scatter Plots<h2>

A pairplot shows scatter plots for all pairs of numerical features and histograms on the diagonal. We use the hue parameter to color the data points by species, which helps us see if the species are separable.

```python
# Create a pairplot to visualize relationships between features
# 'hue' colors the points by the 'species' column, which is incredibly useful
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pair-wise Relationships in the Iris Dataset', y=1.02) # Add a title
plt.show()
```

<h2>5b. Histograms for Feature Distributions<h2>

Histograms help us see the frequency distribution of each numerical feature. The kde (Kernel Density Estimate) line provides a smoothed representation of this distribution.

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Distribution of Iris Features')

# Plot histograms for each numerical feature
sns.histplot(iris_df['sepal_length'], ax=axes[0, 0], kde=True).set_title('Sepal Length')
sns.histplot(iris_df['sepal_width'], ax=axes[0, 1], kde=True).set_title('Sepal Width')
sns.histplot(iris_df['petal_length'], ax=axes[1, 0], kde=True).set_title('Petal Length')
sns.histplot(iris_df['petal_width'], ax=axes[1, 1], kde=True).set_title('Petal Width')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

<h2>5c. Box Plots to Identify Outliers<h2>

Box plots are excellent for summarizing the distribution of each feature across the different species. They clearly show the median, quartiles, and potential outliers (points outside the whiskers).


```python
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Box Plots of Iris Features by Species')

# Plot box plots for each numerical feature, separated by species
sns.boxplot(x='species', y='sepal_length', data=iris_df, ax=axes[0, 0])
sns.boxplot(x='species', y='sepal_width', data=iris_df, ax=axes[0, 1])
sns.boxplot(x='species', y='petal_length', data=iris_df, ax=axes[1, 0])
sns.boxplot(x='species', y='petal_width', data=iris_df, ax=axes[1, 1])

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
